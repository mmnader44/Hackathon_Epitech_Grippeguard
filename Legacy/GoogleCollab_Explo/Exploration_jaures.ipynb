```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Style global
pd.set_option("display.max_columns", None)
plt.style.use("ggplot")

print(" Environnement prêt")
```

```python
print("Colonnes couverture_2024 :", couverture_2024.columns.tolist())
print("Colonnes urgences_reg :", urgences_reg.columns.tolist())
```

```python
urgences_reg.columns = [col.strip().lower().replace(" ", "_") for col in urgences_reg.columns]

urgences_reg.head(2)
```

```python
data_corr = couverture_2024.merge(
    urgences_reg,
    left_on="code",
    right_on="région_code",
    how="inner"
)
print("Fusion réussie :", data_corr.shape)
data_corr.head()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    data=data_corr,
    x="valeur",
    y="taux_de_passages_aux_urgences_pour_grippe",
    hue="groupe"
)
plt.title("Lien entre couverture vaccinale et passages aux urgences (régions)")
plt.xlabel("Taux de couverture vaccinale")
plt.ylabel("Taux de passages aux urgences pour grippe")
plt.show()
```

```python
corr = data_corr[["valeur", "taux_de_passages_aux_urgences_pour_grippe"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Corrélation entre couverture vaccinale et urgences (régions)")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lmplot(
    data=data_corr,
    x="valeur",
    y="taux_de_passages_aux_urgences_pour_grippe",
    hue="groupe",
    height=6,
    aspect=1.3,
    markers=["o", "s"],
    scatter_kws={"s":70, "alpha":0.7},
    line_kws={"lw":2}
)
plt.title("Corrélation entre couverture vaccinale et urgences selon le groupe d'âge (2024)")
plt.xlabel("Taux de couverture vaccinale (%)")
plt.ylabel("Taux de passages aux urgences pour grippe")
plt.show()
```

```python
corr_groupes = (
    data_corr.groupby("groupe")[["valeur", "taux_de_passages_aux_urgences_pour_grippe"]]
    .corr()
    .iloc[0::2, -1]  # On garde la corrélation entre les deux colonnes
    .reset_index()
    .rename(columns={"taux_de_passages_aux_urgences_pour_grippe": "corrélation"})
)

print("Corrélation par groupe d’âge :")
display(corr_groupes)
```

```python
moyennes = (
    data_corr.groupby("groupe")[["valeur", "taux_de_passages_aux_urgences_pour_grippe"]]
    .mean()
    .reset_index()
)

fig, ax1 = plt.subplots(figsize=(8,5))

sns.barplot(data=moyennes, x="groupe", y="valeur", ax=ax1, color="skyblue")
ax1.set_ylabel("Taux de couverture (%)", color="blue")
ax1.set_xlabel("Groupe d'âge")
ax1.tick_params(axis='y', labelcolor="blue")

ax2 = ax1.twinx()
sns.barplot(data=moyennes, x="groupe", y="taux_de_passages_aux_urgences_pour_grippe", ax=ax2, color="salmon", alpha=0.6)
ax2.set_ylabel("Taux de passages aux urgences", color="red")
ax2.tick_params(axis='y', labelcolor="red")

plt.title("Moyenne couverture vs urgences par groupe d’âge (2024)")
plt.show()
```

```python
# On ajoute la colonne année à chaque dataset d'urgences
urgences_reg_2023 = urgences_reg.copy()
urgences_reg_2023["annee"] = 2023

# Idem pour les couvertures
couverture_2023["annee"] = 2023
couverture_2024["annee"] = 2024

# Fusion multi-années (exemple sur 2023–2024)
fusion = pd.concat([couverture_2023, couverture_2024])
fusion = fusion.merge(urgences_reg, left_on="code", right_on="région_code", how="inner")

# Corrélation par année et groupe
corr_multi = (
    fusion.groupby(["annee", "groupe"])[["valeur", "taux_de_passages_aux_urgences_pour_grippe"]]
    .corr()
    .iloc[0::2, -1]
    .reset_index()
    .rename(columns={"taux_de_passages_aux_urgences_pour_grippe": "corrélation"})
)

display(corr_multi)
```
